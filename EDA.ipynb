```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
facebook = pd.read_excel("Final_facebook_with manual impact.xlsx", sheet_name='final_selected')
facebook_num = facebook[['Total Reactions', 'Shares', 'Comments', 'Page_likes','Estimated impact Anna']]
```

```python
insta = pd.read_excel("Instagram_dataset.xlsx", sheet_name='Sheet1')
insta_num = insta[['followers', 'likes', 'views', 'comments','estimated impact anna']]
```

```python
twitter = pd.read_excel("Twitter_NEW.xlsx", sheet_name='Sheet2')
twitter_num= twitter[['favorite_count', 'retweet_count', 'user.followers_count', 'comments','Estimated manual impact (anna)']]
```

```python
#facebookEDA
sns.countplot(x='Estimated impact Anna', data=facebook)
# or sns.distplot(facebook['Estimated impact Anna'], color='g', bins=18, hist_kws={'alpha': 0.5});

#impact per mention_type
sns.factorplot(x='Estimated impact Anna', col='Mention_type', kind='count', data=facebook)

# Count plot to show the count per mention type
sns.set_style("white")
g = sns.factorplot(x="Mention_type", data=facebook, kind="count",
                   palette="BuPu", size=6, aspect=1.5)
g.ax.xaxis.set_label_text("Mention_type",fontdict= {'size':16})
g.ax.yaxis.set_label_text("count", fontdict= {'size':16})
g.ax.set_title("Total count of posts per mention type",fontdict= {'size':18})

#model_var
facebook_num.hist(figsize=(16, 20), xlabelsize=8, ylabelsize=8);
```

```python
#InstagramEDA
sns.countplot(x='estimated impact anna', data=insta)
# or sns.distplot(insta['estimated impact anna'], color='g', bins=18, hist_kws={'alpha': 0.5});

#impact per mention_type
sns.factorplot(x='estimated impact anna', col='type', kind='count', data=insta)

# Count plot to show the count per mention type
sns.set_style("white")
g = sns.factorplot(x="type", data=insta, kind="count",
                   palette="BuPu", size=6, aspect=1.5)
g.ax.xaxis.set_label_text("Mention_type",fontdict= {'size':16})
g.ax.yaxis.set_label_text("count", fontdict= {'size':16})
g.ax.set_title("Total count of posts per mention type",fontdict= {'size':18})

#model_var
insta_num.hist(figsize=(16, 20), xlabelsize=8, ylabelsize=8);
```

```python
#TwitterEDA
sns.countplot(x='Estimated manual impact (anna)', data=twitter)
# or sns.distplot(insta['estimated impact anna'], color='g', bins=18, hist_kws={'alpha': 0.5});


#model_var
twitter_num.hist(figsize=(16, 20), xlabelsize=8, ylabelsize=8);
```
